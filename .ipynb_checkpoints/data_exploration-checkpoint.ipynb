```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
from upload_overleaf.upload import upload
```

```python
# Scan parquet file
#lf_daily = pl.scan_parquet("prices_daily.parquet")
lf_intraday = pl.scan_parquet("prices.parquet")
#lf_coacs = pl.scan_parquet("coacs.parquet", low_memory=True)
```

```python
df_intraday_aapl = (lf_intraday.filter(pl.col("ticker") == "AAPL"))
```

```python
# Plot time series of StockClose
plt.figure(figsize=(20,10))
sns.lineplot(data=df_intraday_aapl, x="date", y="Close")
plt.title("AAPL")
# upload(plt, "Master's Thesis", 'figures/aapl_test.png')
```
